```python
import re
from Bio import Entrez, SeqIO

# Set your email address for NCBI Entrez (required)
Entrez.email = "your_email@example.com"

# Specify the input file containing protein sequences
input_file = "/Users/jra/Desktop/protein_sequences.fasta"  # Replace with your input file name

# Specify the output file for storing results
output_file = "/Users/jra/Desktop/output_sequences.fasta"  # Replace with your output file name

# Initialize an empty list to store protein sequences
protein_sequences = []

# Read protein sequences from the input file
with open(input_file, "r") as infile:
    for line in infile:
        line = line.strip()
        if line.startswith(">"):
            protein_sequences.append(line)

# Initialize an empty list to store results
results = []

# Process each protein sequence
for protein_sequence in protein_sequences:
    try:
        # Use regular expressions to extract accession, taxid, and gene name
        accession_match = re.search(r'\|([^|]+)\|', protein_sequence)
        taxid_match = re.search(r'OX=(\d+)', protein_sequence)
        gene_name_match = re.search(r'GN=([^ ]+)', protein_sequence)

        if accession_match and taxid_match and gene_name_match:
            accession = accession_match.group(1)
            taxid = taxid_match.group(1)
            gene_name = gene_name_match.group(1)

            # Search for the CDS using the derived accession, taxid, and gene name
            search_term = f"{accession} OR {gene_name} AND txid{taxid}[Organism]"
            handle = Entrez.esearch(db="nucleotide", term=search_term, retmax=1)
            record = Entrez.read(handle)

            if "IdList" in record and record["IdList"]:
                # Get the CDS record
                cds_id = record["IdList"][0]
                cds_handle = Entrez.efetch(db="nucleotide", id=cds_id, rettype="gb", retmode="text")
                cds_record = SeqIO.read(cds_handle, "genbank")

                # Store the original protein sequence and CDS sequence in the results list
                results.append(f"{protein_sequence}\n{cds_record.seq}\n")

            else:
                print(f"No CDS found for Protein Sequence: {protein_sequence}")

        else:
            print(f"Unable to extract required information from Protein Sequence: {protein_sequence}")

    except Exception as e:
        print(f"Error processing Protein Sequence: {protein_sequence}")
        print(e)

# Write the results to the output file
with open(output_file, "w") as outfile:
    outfile.write("\n".join(results))

print(f"Results saved to {output_file}")
```
